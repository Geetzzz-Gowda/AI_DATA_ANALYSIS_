```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/workspaces/AI_DATA_ANALYSIS_/src/Module 2/grouped_mean_airtravel.csv')

df.columns = df.columns.str.strip().str.replace('"', '').str.replace(" ", "")

print(df.head())

df['Month'] = pd.to_datetime(df['Month'], format='%b').dt.month

df = df.sort_values('Month')

df['Month'] = df['Month'].apply(lambda x: pd.to_datetime(str(x), format='%m').strftime('%b'))

plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Month', y='1958', marker='o', label='1958')
sns.lineplot(data=df, x='Month', y='1959', marker='o', label='1959')
sns.lineplot(data=df, x='Month', y='1960', marker='o', label='1960')

plt.title('Monthly Air Travel (1958-1960)')
plt.ylabel('Number of Passengers (in 1000s)')
plt.xlabel('Month')
plt.legend()
plt.grid(True)
plt.show()
```
