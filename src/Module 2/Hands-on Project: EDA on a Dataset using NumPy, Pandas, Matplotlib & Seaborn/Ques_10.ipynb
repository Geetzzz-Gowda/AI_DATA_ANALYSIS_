```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def clean_and_prepare_data(file_path):
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("Error: File not found. Please check the file path.")
        return None
    except pd.errors.ParserError:
        print("Error: Could not parse the CSV file. Please check its format.")
        return None

    df.columns = df.columns.str.strip().str.replace('"', '').str.replace(" ", "")
    df['Month'] = pd.to_datetime(df['Month'], format='%b').dt.month
    df = df.sort_values('Month')
    df['Month'] = df['Month'].apply(lambda x: pd.to_datetime(str(x), format='%m').strftime('%b'))
    return df

def plot_monthly_air_travel(df):
    plt.figure(figsize=(10, 6))
    for year in ['1958', '1959', '1960']:
        sns.lineplot(data=df, x='Month', y=year, marker='o', label=year)

    plt.title('Monthly Air Travel (1958-1960)')
    plt.ylabel('Number of Passengers (in 1000s)')
    plt.xlabel('Month')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

file_path = '/workspaces/AI_DATA_ANALYSIS_/src/Module 2/grouped_mean_airtravel.csv'
df = clean_and_prepare_data(file_path)

if df is not None:
    plot_monthly_air_travel(df)
```
