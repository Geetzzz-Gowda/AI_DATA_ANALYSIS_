```python
# Basic Seaborn Bar Plot
# Description: Using Seaborn, plot the same bar plot from Question 1 to compare the sales data.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Sales': [120, 340, 230, 150, 300]
}

df = pd.DataFrame(data)

sns.barplot(x='Product', y='Sales', data=df, palette='pastel')

plt.title('Sales Comparison by Product')
plt.xlabel('Product')
plt.ylabel('Sales')

plt.show()
```
