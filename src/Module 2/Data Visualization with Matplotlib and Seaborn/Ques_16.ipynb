```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
```

```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D', 'E']
sales = [120, 340, 230, 150, 300]

height = [160, 170, 165, 180, 175]
weight = [55, 65, 60, 80, 76]

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].bar(products, sales, color='skyblue')
axs[0].set_title('Product Sales')
axs[0].set_xlabel('Product')
axs[0].set_ylabel('Sales')

axs[1].scatter(height, weight, color='orange', edgecolor='black')
axs[1].set_title('Height vs Weight')
axs[1].set_xlabel('Height (cm)')
axs[1].set_ylabel('Weight (kg)')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
