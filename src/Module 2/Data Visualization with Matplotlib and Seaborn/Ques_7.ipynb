```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def plot_scatter(
    data,
    x_col='Height',
    y_col='Weight',
    size_col='Size',
    color='royalblue',
    size_range=(50, 300),
    title='Scatter Plot',
    xlabel=None,
    ylabel=None,
    show_legend=True
):
    # Validation
    if not isinstance(data, pd.DataFrame):
        raise TypeError("Input must be a pandas DataFrame.")
    
    for col in [x_col, y_col, size_col]:
        if col not in data.columns:
            raise ValueError(f"Missing required column: '{col}'")
        if not pd.api.types.is_numeric_dtype(data[col]):
            raise TypeError(f"Column '{col}' must be numeric.")

    # Plot
    sns.set(style='whitegrid')
    plot = sns.scatterplot(
        x=x_col,
        y=y_col,
        size=size_col,
        data=data,
        sizes=size_range,
        color=color,
        legend=show_legend
    )

    plot.set_title(title, fontsize=14, fontweight='bold')
    plot.set_xlabel(xlabel if xlabel else x_col)
    plot.set_ylabel(ylabel if ylabel else y_col)

    if show_legend:
        handles, labels = plot.get_legend_handles_labels()
        if size_col in labels:
            plot.legend(title=size_col, loc='upper left', bbox_to_anchor=(1, 1))
        else:
            plot.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.tight_layout()
    plt.show()

# Example usage
data = pd.DataFrame({
    'Height': [160, 170, 165, 180, 175],
    'Weight': [55, 65, 60, 80, 76],
    'Size': [10, 15, 13, 18, 12]
})

plot_scatter(
    data,
    x_col='Height',
    y_col='Weight',
    size_col='Size',
    color='seagreen',
    size_range=(50, 300),
    title='Height vs Weight (Sized by Data)',
    xlabel='Height (cm)',
    ylabel='Weight (kg)',
    show_legend=True
)
```
