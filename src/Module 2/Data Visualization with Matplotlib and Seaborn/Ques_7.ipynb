```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def plot_height_weight_scatter(data, color='royalblue', size_range=(50, 300)):
    required_columns = ['Height', 'Weight', 'Size']
    
    # Validation
    if not isinstance(data, pd.DataFrame):
        raise TypeError("Input data must be a pandas DataFrame.")
    
    for col in required_columns:
        if col not in data.columns:
            raise ValueError(f"Missing required column: '{col}'")
        if not pd.api.types.is_numeric_dtype(data[col]):
            raise TypeError(f"Column '{col}' must be numeric.")

    # Plotting
    sns.scatterplot(
        x='Height',
        y='Weight',
        size='Size',
        data=data,
        sizes=size_range,
        color=color,
        legend=False
    )
    
    plt.title('Height vs Weight (Sized by Data)')
    plt.xlabel('Height (cm)')
    plt.ylabel('Weight (kg)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# Sample data
data = pd.DataFrame({
    'Height': [160, 170, 165, 180, 175],
    'Weight': [55, 65, 60, 80, 76],
    'Size': [10, 15, 13, 18, 12]
})

# Call the function
plot_height_weight_scatter(data, color='seagreen', size_range=(40, 250))
```
