```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D', 'E']
sales = [120, 340, 230, 150, 300]

plt.bar(products, sales, color='mediumseagreen')

plt.title('Sales Comparison by Product', fontsize=16, fontweight='bold')
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sales', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
