```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(42)
data = pd.DataFrame(np.random.randn(10, 6), columns=list('ABCDEF'))

# Compute correlation matrix
corr = data.corr()

# Plot heatmap with advanced styling
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,               # Show correlation values
    fmt=".2f",                # Format to 2 decimal places
    cmap='coolwarm',          # Color map
    linewidths=0.5,           # Lines between cells
    linecolor='gray',
    cbar_kws={'shrink': 0.8}, # Colorbar size
    square=True,
    annot_kws={'size': 10, 'weight': 'bold', 'color': 'black'}
)

plt.title('Correlation Matrix Heatmap', fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```
